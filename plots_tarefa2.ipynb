```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
cat_bots = [('Baby', 7819), ('PC', 34630), ('Personal_Care_Appliances', 603), ('Pet Products', 15877), 
            ('Jewelry', 7825), ('Mobile_Electronics', 452), ('Music', 258477), ('Toys', 42652), 
            ('Video DVD', 275148), ('Beauty', 41400), ('Apparel', 17373), ('Digital_Music_Purchase', 26950),
            ('Outdoors', 9829), ('Home Improvement', 16530), ('Health & Personal Care', 40694), 
            ('Office Products', 22458), ('Home', 40086), ('Video', 29282), ('Tools', 9112), 
            ('Video Games', 17897), ('Musical Instruments', 4004), ('Luggage', 1781), ('Software', 5777), 
            ('Books', 945457), ('Kitchen', 38719), ('Wireless', 31585), ('Furniture', 2787), ('Automotive', 8500), 
            ('Electronics', 18975), ('Digital_Software', 255), ('Gift Card', 305), ('Grocery', 39597), 
            ('Sports', 18440), ('Camera', 10894), ('Watches', 2941), ('Digital_Video_Games', 901), 
            ('Digital_Video_Download', 25225), ('Home Entertainment', 3494), ('Lawn and Garden', 15718), 
            ('Mobile_Apps', 13940), ('Digital_Ebook_Purchase', 420044), ('Shoes', 8304), 
            ('Major Appliances', 427)]

print(len(cat_bots))

labels, ys = zip(*cat_bots)
xs = np.arange(len(labels)) 
width = 1

fig = plt.figure(figsize=(40,20))                                                               
ax = fig.gca()  #get current axes
ax.bar(xs, ys, width, align='center')

#Remove the default x-axis tick numbers and  
#use tick numbers of your own choosing:
ax.set_xticks(xs)
#Replace the tick numbers with strings:
ax.set_xticklabels(labels)
#Remove the default y-axis tick numbers and  
#use tick numbers of your own choosing:
ax.set_yticks(ys)

plt.savefig('netscore.png')
```

```python
fig = plt.figure(figsize=(40,20))                                                               
ax = fig.gca()  #get current axes
mylabels = [x[0] for x in cat_bots]
print(mylabels)
patches, texts = ax.pie( ys)
plt.legend(patches, mylabels, loc="best")
plt.savefig('netscore.png')

```
